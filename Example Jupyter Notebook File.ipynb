## Example Code featuring Normal Distibution

(example based on https://arachnoid.com/IPython/normal_distribution.html)

Probability density function of the normal distribution:

$$
f(x|\mu,\sigma^2)= \frac{1}{\sigma\sqrt{2 \pi}} \cdot e^{-\frac{(x-\mu)^2}{2\sigma^2}\cdot }
$$


````python
import numpy as np
from sympy import *
from IPython.display import *
init_printing(use_latex=True)
import matplotlib.pyplot as plt
%matplotlib inline
fig = plt.gcf()
fig.set_size_inches(8,5)
var('a b x sigma mu')
pdf = Lambda((x,mu,sigma),
  (1/(sigma * sqrt(2*pi)) * exp(-(mu-x)**2 / (2*sigma**2)))
)
ltx = '$ f(x|\\mu,\\sigma^2) = \\frac{1}{ \\sigma' + \
 '\\sqrt{2 \\pi}} e^{\\left(-\\frac{{\\left(\\mu - ' + \
 'x\\right)}^{2}}{2 \\, \\sigma^{2}}\\right)}$'
display(Latex(ltx))
x = np.linspace(50,150,100)
y = np.array([pdf(v,100,15) for v in x],dtype='float')
plt.grid(True)
plt.title('Probability Density Function')
plt.xlabel('x')
plt.ylabel('$ f(x|\\mu,\\sigma^2) $')
plt.text(122,0.023,'$\mu = 100$',fontsize=16)
plt.text(122,0.021,'$\sigma = 15$',fontsize=16)
plt.plot(x,y,color='gray')
plt.fill_between(x,y,0,color='#c0f0c0')
plt.show()

````
